# Orthonormal basis for $L_2(\mathbb{R}^n, \mu)$

## Hermite polynomials 

- $\mu = $ Gaussian measure
- $k = RBF$ 

```python
def exp2(x):
    y = np.exp(-x**2)
    return(y)

def hPoly(x, k):
    gradF = exp2
    for i in np.arange(1, k+1):
        gradF = egrad(gradF)
    y = (-1)**k *  np.exp(x**2) * gradF(x)
    return(y)
```

```python
x = np.arange(-2, 4, 0.01)

order = 4
for k in range(1, order + 1):
    y = hPoly(x, k)
    plt.plot(x, y)
    
plt.ylim(-25, 20)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.show()
```

## Finite spectral density

$$
k_{\lambda}^N(x, y) = \sum_{i = 1}^N \lambda_i \phi_i(x) \ \phi_i(y)
$$

The process:

$$
f_{\lambda}^N(x) = \sum_{i = 1}^N (\lambda_i)^{1/2} \phi_i(x) \ \omega_i
$$

where:

$$
\omega_i \sim \mathcal{N}(0, \sigma^2), \ iid.
$$

```python
def eValRFB(k, s, l):
    v = s**2
    l2 = l**2
    
    a = 1/(4*v)
    b = 1/(2*l2)
    c = np.sqrt(a**2 +2*a*b)
    
    A = a + b + c
    B = b/A
    
    L = np.sqrt(2*a/A) * B**k
    
    return(L)
    
```

```python
def eFunRFB(x, k, s, l):
    v = s**2
    l2 = l**2
    
    a = 1/(4*v)
    b = 1/(2*l2)
    c = np.sqrt(a**2 + 2*a*b)
    
    e = np.exp(-(c - a) * (x**2))
    h = hPoly(np.sqrt(2*c)*x, k)
    
    return(e * h)
```

```python
x = np.arange(-6, 6, 0.01)
npts = x.shape[0]

l = 0.3
s = 5

y0 = eFunRFB(x, 0, s, l)
y1 = eFunRFB(x, 1, s, l)
y2 = eFunRFB(x, 2, s, l)
y3 = eFunRFB(x, 3, s, l)
```

```python
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.title('Eigenfunctions of for the gaussian measures and RBF kernel')
plt.show()
```

```python
nEigVals = 100
mu = np.zeros((nEigVals))
K = np.eye(nEigVals)

wR = np.random.multivariate_normal(mu, K, 1)
w = np.transpose(wR)
```

```python
phiMat = np.zeros((npts, nEigVals))

for k in np.arange(0, nEigVals):
    y = eFunRFB(x, k, s, l)
    phiMat[:, k] = y
    
```

```python
print(phiMat.shape, w.shape)
f = np.dot(phiMat, w)

plt.plot(x, f)
plt.show()
```
